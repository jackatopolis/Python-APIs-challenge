```python
cities = []
print("Beginning Data Retrieval")
print("------------------------")
for i in range(len(lats)):
    city_info = cp.nearest_city(lats[i],longs[i])
    url = f'https://api.openweathermap.org/data/2.5/weather?q={city_info.city_name},{city_info.country_code}&appid={weather_key}'
    try: 
        data = req.get(url).json()
        cities.append({'city':city_info.city_name,
                       'country':city_info.country_code.upper(),
                       'lat':lats[i],'lon':longs[i],
                       'temp':data['main']['temp'],
                       'temp_min':data['main']['temp_min'],
                       'temp_max':data['main']['temp_max'],
                       'pressure':data['main']['pressure'],
                       'humidity':data['main']['humidity'],
                       'cloudiness':data['clouds']['all'],
                       'wind_speed':data['wind']['speed'],
                       'timestamp':data['dt']
                       })
        pringt(f'Processing City {i} | {city_info.city_name}, {city_info.country_code}')
    except Exception as e:
        print(f'City not found. Skipping...')
print("------------------------")
print("Data Retrieval Complete")
print("------------------------")
```
